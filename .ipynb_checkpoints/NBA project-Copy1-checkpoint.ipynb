```python
import pandas as pd
import csv
import matplotlib.pyplot as plt

closed_file = 'NBA PROJECTS.csv'

nba = pd.read_csv(closed_file)

nba = nba.loc[(nba['Tm'] != 'TOT')& (nba['Year'] >= 1990)] 
nba
```

```python
nba.columns
```

```python
threes = nba.groupby('Year').sum()['3PA']
threeline = threes.plot(kind = 'line')
print(threeline)
```

```python
threeteam = nba.groupby('Tm').sum()['WS']
teamsthree = pd.DataFrame(threeteam)

teamsthree.sort_values(by = 'WS', ascending = False)
```

```python
player = input('Which Player would you like to study?')
indplayer = nba.loc[nba['Player']== player]
plt.plot(indplayer['Year'], indplayer['USG%'], color = 'Red')
```
